```python
import seaborn as sns
import pandas as pd
#reference https://seaborn.pydata.org/generated/seaborn.relplot.html
```

```python
sns.set()
diabetes = pd.read_csv('../data/diabetes.csv')
diabetes
```

```python
sns.relplot(
            x="Age", y="Pregnancies",col="Outcome",
            kind="line",
            data=diabetes);
```

```python
sns.set(style="ticks", color_codes=True)
#g = sns.pairplot(diabetes)
```

```python
g = sns.pairplot(diabetes,height=3,
                 vars=["Age", "BMI","Pregnancies"],hue = 'Outcome')
```

```python
g = sns.pairplot(diabetes,height=3,
                 plot_kws={ 'scatter_kws': {'alpha': 0.2}},
                 vars=["Age", "BMI","Pregnancies"],kind="reg",hue = 'Outcome')
```
